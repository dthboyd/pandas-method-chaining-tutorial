## Method chaining
It is common practice and encouraged in the Pandas documentation to use 'method chaining.' I use method chaining, for example, to read in my data and to group/sample/slice before plotting. The advantages of method chaining include:
- It saves you from having to define a lot of intermediate variables.
- You can clean your data directly after you read it in, combining all your pre-processing steps into a single command.
- You can plot your data directly after grouping/sampling/slicing/etc.
- It looks good. Your code will read like a story; top to bottom, left to right.

Or, in Tom Augspurger's words:

> "My scripts will typically start off with large-ish chain at the start getting things into a manageable state. It's good to have the bulk of your munging done with right away so you can start to do Science™"

Source and fantastic walkthrough: http://tomaugspurger.github.io/method-chaining.html (read all 7 parts)

Other tips:
- Use pandas build-in methods when possible, like `resample` below. Otherwise, define your own function and use `pipe`.
- As always, prefer vectorized operations over iterative operations (see `correct_flow_rate_fast` below)
- Familiarize yourself with the difference between `.loc`, `.iloc` and pandas assignment in general ([link](http://tomaugspurger.github.io/modern-1.html))

Resources:
- http://pandas.pydata.org/pandas-docs/stable/generated/pandas.read_csv.html
- http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.pipe.html
- http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.resample.html

```python
# The main method-chain
def read(file_path):
    df = (pd.read_csv(file_path, 
                      sep = '\t', 
                      skiprows = 2,
                      usecols = ['Time','Ground Water Meter Total (L)'])
            .pipe(drop_bad_times)
            .pipe(parse_dates)
            .set_index('Time')
            .resample('5T').mean()
            .pipe(calc_flow_rate)
            .pipe(correct_flow_rate_fast))
            #.apply(correct_flow_rate_slow, axis = 1))
    
    return df

# Auxiliary functions
def drop_bad_times(df):
    df = df.copy()
    df = df[df['Time'] != 'LOADOK']
    return df

def parse_dates(df):
    df = df.copy()
    df['Time'] = pd.to_datetime(df['Time'], format = '%d/%m/%Y %H.%M.%S')
    return df

def calc_flow_rate(df):
    df = df.copy()
    df['Delta_time'] = (df.index[1] - df.index[0]).seconds
    df['Flow']       = (df['Ground Water Meter Total (L)'].diff(1) / df['Delta_time']).fillna(0)
    return df

def correct_flow_rate_slow(row):
    if row['Flow'] < -10:
        row['Flow'] = row['Flow'] + 65535 / row['Delta_time']
    return row

def correct_flow_rate_fast(df):
    df   = df.copy()
    crit = (df['Flow'] < -10)
    df.loc[crit, 'Flow'] = df.loc[crit, 'Flow'] + 65535 / df.loc[crit, 'Delta_time']
    return df
  
t       = time.time()
df      = read('data/raw_data.txt')
elapsed = time.time() - t

print("time elapsed =", elapsed)
```

In this example, we are using data from a water meter. The data is in fairly good shape, but does require some pre-processing and cleaning. The 'method chain' in the `read` function above consists of the following steps:

1. `read_csv` to read in the csv-file, skip the first two rows and select two columns
2. `pipe(drop_bad_times)` to filter out rows with invalid measurements
3. `pipe(parse_dates)` to parse the date column to a pandas datetime format
4. `set_index('Time')` to set the date/time as the dataframe's index
5. `resample('5T').mean()` to resample and interpolate the data at standard 5 minute intervals
6. `pipe(calc_flow_rate)` to compute a new variable, in this case the flow rate
7. `pipe(correct_flow_rate_{fast,slow}))` to correct the flow rate. Note that the vectorized operation is much quicker than `apply`.

Overall, this creates a nice story of reading in the data and applying some pre-processing steps in one go. In practice, I find that many of the operations mentioned here (such as parsing the dates and creating new columns) are required for nearly every csv file.

### Visualization

`matplotlib` will do nearly anything you want, but it can be cumberstone to use. I tend to use the `pandas` build-in `plot` function regularly, in addition to the `seaborn` plotting functionality. This blog by Dan Saber is really useful (and hilarious) to compare what's out there: https://dansaber.wordpress.com/2016/10/02/a-dramatic-tour-through-pythons-data-visualization-landscape-including-ggplot-and-altair/

```python
# Load some prepared data
with open('data/processed_data', 'rb') as input_file:
    dg = pickle.load(input_file)
    
dg['LWT G'] = dg[['T in 1','T in 2']].mean(axis=1)
dg['EWT G'] = dg[['T out 1','T out 2']].mean(axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
```

Using `Seaborn`'s `set_style` and `set_context` is an easy way to manage the initial layout. I often use `set_style("white")` and `set_style("ticks")`.

Source: https://stanford.edu/~mwaskom/software/seaborn/tutorial/aesthetics.html

```python
sns.set_style("white")
sns.set_context("talk")
```

`matplotlib` is flexible but cumberstone, especially when it comes to plotting multiple groups of data on the same graph.

```python
fig, ax = plt.subplots(1, 1, figsize=(16,5), dpi=300)

# Have to repeat/loop this:
ax.plot(dg.loc[:,'Date Time'],
        dg.loc[:,'LWT G'],
        label = 'Exiting water temperature')

ax.plot(dg.loc[:,'Date Time'],
        dg.loc[:,'EWT G'],
        label = 'Entering water temperature')

ax.set_xlabel('')
ax.set_ylabel('Water temperature ($^\circ$C)')
ax.set_title('Water Temperature entering and exiting the GSHP system')
ax.legend()

sns.despine(ax=ax)
fig.savefig('temperature.png', format='png', dpi=500)
```

Using `pandas`, plotting multiple columns is a lot more convenient. Using the format below (calling `.plot` with `ax = ax`) you also get the usability of `pandas` and the flexibility of `matplotlib` to adjust the figure aesthetics.

```python
fig, ax = plt.subplots(1, 1, figsize=(16,5), dpi=300)

dg.plot(x      = 'Date Time',
        y      = ['LWT G','EWT G'],
        kind   = 'line', 
        legend = False,
        ax     = ax)

ax.set_xlabel('')
ax.set_ylabel('Water temperature ($^\circ$C)')
ax.set_title('Water Temperature entering and exiting the GSHP system')
ax.legend(['Exiting water temperature','Entering water temperature'])

sns.despine(ax=ax)
```

### Combining method chaining and pandas plotting
To bring it all together: because you can call `.plot` on a dataframe, we can include it as a method in our chain. 

This creates a nice flow of slicing, grouping and plotting in the example below.

```python
fig, ax = plt.subplots(1, 1, figsize=(16,5), dpi=300)

(dg.loc[(dg['Date Time'] >= pd.datetime(2016,1,1)) &
        (dg['Date Time'] <  pd.datetime(2016,10,1))] 
   .groupby(dg['Date Time'].dt.hour) 
   .mean()
   .plot(y      = 'Power consumed',
         kind   = 'line', 
         legend = False,
         ax     = ax))

ax.set_xlabel('Hour of the day')
ax.set_ylabel('Average power (kW)')
ax.set_title('Average power consumed by the GSHP system, per hour')

sns.despine(ax=ax)
```
